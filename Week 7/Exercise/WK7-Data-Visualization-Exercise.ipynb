```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"..\raw\chip-dataset-cleaned.csv",
                 delimiter=",")
df.info()
```

```python
df = df.drop(df.columns[[10,11,12]],axis=1)
df.head()
```

# N Chips Manufactured

```python
df_w = df.copy()
```

```python
chip_count = df_w['Vendor'].value_counts()
chip_count
```

```python
type(chip_count)
```

```python
chart = chip_count.plot(kind = 'bar',
                  figsize=(10,5),
                  color = ['#FFDAB3','#C8AAAA','#9F8383','#574964','#212121'])
chart.bar_label(chart.containers[0], label_type='edge')
plt.title('Chips Manufactured (CPU+GPU)',
          fontfamily = 'Palatino Linotype',
          fontsize=12,
          fontweight='bold')
plt.xlabel('')
chart.set_xticklabels(chip_count.index, rotation=0)
plt.ylabel('N Chips')
```

# N CPU Manufactured

```python
df_cpu = df_w[df_w['Type']=='CPU']
df_cpu
```

```python
cpu_count = df_cpu['Vendor'].value_counts()
cpu_count
```

```python
chart = cpu_count.plot(kind='bar',
                        figsize=(6,3),
                        color = ['#FFDAB3','#C8AAAA'])
plt.title('CPU Manufactured')
plt.ylabel('N Count')
plt.xlabel('')
plt.show()
```

# Lowest TDP (Thermal Design Power)

```python
df_tdp = df_w[['Product','TDP']]
df_tdp
```

```python
type(df_tdp)
```

```python
df_tdp_ascending = df_tdp.copy()
df_tdp_ascending = df_tdp_ascending.sort_values(by='TDP', ascending = True)
df_tdp_ascending
```

```python
df_tdp_ascending = df_tdp_ascending.head(7)
df_tdp_ascending
```

```python
plt.figure(figsize=(10,5))
plt.bar(df_tdp_ascending['Product'],df_tdp_ascending['TDP'],
        color = ['#34073d','#531942','#722b47','#923e4d','#b15052','#d06257','#ef745c'])
plt.title('Products with the Lowest TDP')
plt.ylabel('Watts')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
df_tdp_descending = df_tdp.copy()
df_tdp_descending = df_tdp_descending.sort_values(by='TDP', ascending = False)
df_tdp_descending
```

```python
df_tdp_descending = df_tdp_descending.head(7)
df_tdp_descending
```

```python
plt.figure(figsize=(10,5))
plt.bar(df_tdp_descending['Product'],df_tdp_descending['TDP'],
        color = ['#ef745c','#d06257','#b15052','#923e4d','#722b47','#531942','#34073d'])
plt.title('Products with the Highest TDP')
plt.ylabel('Watts')
plt.xticks(rotation=45, ha='right')
plt.show()
```

# GPU Process Size Trend

```python
df_ps = df_w[['Type','Date','Process_Size']].where(df_w['Type']=='GPU').dropna()
df_ps
```

```python
df_ps['Date'] = pd.to_datetime(df_ps['Date'])
df_ps['Year'] = df_ps['Date'].dt.year
df_ps 
```

```python
df_ps = df_ps.groupby('Year')['Process_Size'].mean().reset_index()
df_ps
```

```python
plt.figure(figsize=(10,5))
plt.plot(df_ps['Year'],df_ps['Process_Size'])
plt.title('GPU Process Size Trend')
plt.ylabel('nm')
```
